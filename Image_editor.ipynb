```python
import PySimpleGUI as sg

import PIL

import Imagefunctions as fn

import imutils

import cv2

import os

path = "butterfly.jpg"
temp = cv2.imread(path)
cv2.imwrite('butterfly.png',temp)
savepath="give image save path here"
```

```python
sg.theme('Black')   # Add a touch of color
# All the stuff inside your window.

def make_win():
   
    col1 = [
            
        [sg.Text(text = 'select image using browse', key = '-filename-'),sg.FileBrowse(file_types = (("jpg", "*.jpg"), ("PNG", "*.png"),),key='-path-')],
            
        [sg.Button('Load Image', key = '-load-',border_width = 8,)],
           
        [sg.Button('Blackandwhite',border_width=8,)], 
           
        [sg.Button('Negative',border_width=8)], 
          
        [sg.Button('Flip',border_width=8)],
           
        [sg.Button(key = 1,image_filename = 'clockwise.png',tooltip = 'Rotate clockwise',border_width=10,image_subsample=2,),
             
         sg.Button(key = 2,image_filename = 'anticlockwise.png',tooltip = 'Rotate anticlockwise',border_width=10,image_subsample=2)],
           
        [sg.Button('Edge Dtetection', border_width=8,)],
            
        [sg.Input(key = "-width-",size = (10,10),default_text=""),sg.Button('Resize',key = 'resize',border_width=10,tooltip="enter value of width in the box and press")],
           
        [sg.Button('Undo',key='undo',border_width=8)],
        
        [sg.Input('Name of Image',key='-name-',size=(20,10)),sg.Button('SAVE',key='-save-',border_width=8)],
            
          
            
]
    col2 = [
    
        [sg.Image(key='image',filename="Butterfly.png")]
]

    layout = [
    
        [sg.Column(col1,element_justification='centre'),sg.VSeperator(),sg.Column(col2,element_justification='centre')]
    ] 
    
    
         
    return sg.Window("Image Editor by Vedmani Vaidya",layout,resizable=True,)
    
         
```

```python
# Create the Window
window = make_win()
# Event Loop to process "events" and get the "values" of the inputs
while True:             # Event Loop
    event, values = window.Read()
    
    if event == sg.WIN_CLOSED:
        
        break
    
    
    if event == '-load-':
        
        try:
            
            filename = values['-path-']
            
            if os.path.exists(filename):
                
                path = filename
                
                window['-filename-'].update(value = path ,background_color = None,text_color = None,font = None,visible = None)
                
                temp = fn.loadimage(path)
                
                undo = temp
                
                window['image'].update(filename = 'final.png',data = None,size = (None, None),visible = None)
            
            else:
                
                sg.Popup('invalid file')
                
        
        except:
            
            sg.Popup('error')
            
    
    if event == 'Blackandwhite':
       
        try:
            
            undo = temp
            
            temp = fn.blackandwhite(temp)
            
            window['image'].update(filename = 'final.png',data = None,size = (None, None),visible = None)
            
               
        except:
            
            sg.Popup('Gray image can not be coverted again') 
    
    if event == 'Negative':
       
        try:
            
            undo = temp
            
            temp = fn.negative(temp)
            
            window['image'].update(filename = 'final.png',data = None,size = (None, None),visible = None)
       
        except:
            
            sg.Popup('error')
    
    if event == 'Flip':
        
        try:
            
            undo = temp
            
            temp = fn.flip(temp)
            
            window['image'].update(filename = 'final.png',data = None,size = (None, None),visible = None)
        
        except:
            
            sg.Popup('error')
    
    if event == 1:
        
        try:
            
            undo = temp
            
            temp = fn.anticlockwise_rotate(temp)
            
            window['image'].update(filename = 'final.png',data = None,size = (None, None),visible = None)
        
        except:
            
            sg.Popup('error')
   
    if event == 2:
        
        try:
            
            undo = temp
            
            temp = fn.clockwise_rotate(temp)
            
            window['image'].update(filename = 'final.png',data = None,size = (None, None),visible = None)
        
        except:
            
            sg.Popup('error')
    
    if event == 'Edge Dtetection':
        
        try:
            
            undo = temp
            
            temp = fn.canny_edge(temp)
            
            window['image'].update(filename = 'final.png',data = None,size = (None, None),visible = None)
        
        except:
            
            sg.Popup('No more changes can be done')
    
    if event == 'resize':
        
        try:
            
            undo = temp
            
            width1 = values['-width-']
            
            width = int(width1)
            
            temp = fn.resize(temp,width)
            
            window['image'].update(filename = 'final.png',data = None,size = (None, None),visible = None)
        
        except:
            
            sg.Popup ('error')
            
    if event == 'undo':
        
        try:
            
            fn.undo(undo)
            
            temp=undo
            
            window['image'].update(filename = 'final.png',data = None,size = (None, None),visible = None)
        
        except:
            
            sg.Popup('error')
            
            
    if event == '-save-':
        try:
            imgname = values['-name-']
            savefinalpath = savepath +'/'+ values['-name-'] + ".jpg"
            fn.save(savefinalpath)
            sg.Popup('Image saved')
            
            
            
            
            
        except:
            sg.Popup('something went wrong try again')
            
    
            
    
            
    
    if event == 'close':
        
        break

window.Close()
```
